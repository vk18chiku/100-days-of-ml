```python
def rename(name):
    if name=='Royal Challengers Bangalore':
        return 'RCB'
    elif name=='Indian Premier League':
        return 'IPL'
    else:
        return name
```

```python
rename('Royal Challengers Bangalore')
```

```python
rename('Indian Premier League')
```

#### CONVERTORS

```python
pd.read_csv('IPL.csv',converters={'bowling_team':rename,'event_name':rename})
```
