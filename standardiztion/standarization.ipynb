```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('Social_Network_Ads (1).csv')
df
```

```python
# JO KAM K COLUMN NHI H UNKO PEHLE REMOVE KRO

new_df=df.iloc[:,2:]
new_df
```

#### TRAIN TEST SPLIT

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(new_df.drop('Purchased',axis=1),new_df['Purchased'],test_size=0.3,random_state=0)
```

#### STANDARD SCALER

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
# fit the scaler to the train set, it will learn the parameters

scaler.fit(X_train)

X_train_scaled=scaler.transform(X_train)
X_test_scaled=scaler.transform(X_test)

X_test_scaled
```

```python
X_train_scaled=pd.DataFrame(X_train_scaled,columns=X_train.columns)
X_train_scaled
```

```python
X_test_scaled=pd.DataFrame(X_test_scaled,columns=X_test.columns)
X_test_scaled
```

##### EFFECT OF SCALING

```python
# SCALING K KARAN SARA SAME RANGE M AA GAYE WRNA SALARY OR AGE M BAHUT JADA RANGES K DIFFERNCE THA

fig , (ax1,ax2)=plt.subplots(ncols=2,figsize=(9,4))
ax1.scatter(X_train['Age'],X_train['EstimatedSalary'])
ax1.set_title('Before scaling')

ax2.scatter(X_train_scaled['Age'],X_train_scaled['EstimatedSalary'],color='green')
ax2.set_title('Before scaling')
```

```python
fig , (ax1,ax2)=plt.subplots(ncols=2,figsize=(9,4))

ax1.set_title('Before scaling')
sns.kdeplot(X_train['Age'],ax=ax1)
sns.kdeplot(X_train['EstimatedSalary'],ax=ax1)

ax2.set_title('After scaling')
sns.kdeplot(X_train_scaled['Age'],ax=ax2)
sns.kdeplot(X_train_scaled['EstimatedSalary'],ax=ax2)
```

```python
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(12, 5))

# before scaling
ax1.set_title('Age Distribution Before Scaling')
sns.kdeplot(X_train['Age'], ax=ax1)

# after scaling
ax2.set_title('Age Distribution After Standard Scaling')
sns.kdeplot(X_train_scaled['Age'], ax=ax2)
```

```python
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(12, 5))

# before scaling
ax1.set_title('Salary Distribution Before Scaling')
sns.kdeplot(X_train['EstimatedSalary'], ax=ax1)

# after scaling
ax2.set_title('Salary Distribution Standard Scaling')
sns.kdeplot(X_train_scaled['EstimatedSalary'], ax=ax2)
plt.show()
```

##### WHY SCALING IS IMPORTANT 

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr=LogisticRegression()
lr_scaled=LogisticRegression()
```

```python
lr.fit(X_train,y_train)
lr_scaled.fit(X_train_scaled,y_train)
```

```python
y_pred=lr.predict(X_test)
y_pred
```

```python
y_pred_scaled=lr_scaled.predict(X_test_scaled)
y_pred_scaled
```

```python
from sklearn.metrics import accuracy_score
```

```python
print("Actual",accuracy_score(y_test,y_pred))
print("Scaled",accuracy_score(y_test,y_pred_scaled))
```

```python
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier()
dt_scaled = DecisionTreeClassifier()
```

```python
dt.fit(X_train,y_train)
dt_scaled.fit(X_train_scaled,y_train)
```

```python
y_pred=dt.predict(X_test)
y_pred
```

```python
y_pred_scaled=dt_scaled.predict(X_test_scaled)
y_pred_scaled
```

```python
print('Actual',accuracy_score(y_test,y_pred))
print('Afterscaled',accuracy_score(y_test,y_pred_scaled))
```
